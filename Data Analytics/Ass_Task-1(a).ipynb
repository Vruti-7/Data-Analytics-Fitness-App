# Task-1:Advanced Statistical Analysis for Software Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Task1 a)Analyse the provided dataset to identify relevant features
#and patterns related to user engagement and app performance. 
# Step 1: Data Loading
data = {
    'User ID': [1, 2, 3, 4, 5],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Age': [30, 25, 40, 35, 28],
    'Activity Level': ['Active', 'Moderate', 'Sedentary', 'Active', 'Moderate'],
    'Location': ['Urban', 'Rural', 'Suburban', 'Urban', 'Rural'],
    'App Sessions': [100, 50, 200, 150, 75],
    'Distance Travelled (km)': [50, 20, 100, 75, 30],
    'Calories Burned': [500, 200, 1000, 750, 300]
}
df = pd.DataFrame(data)
```

```python
# Step 2: Data Exploration
print("Dataset Overview:")
print(df.head())
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 3: Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# Step 4: Visualization
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')

plt.subplot(2, 3, 2)
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')

plt.subplot(2, 3, 3)
sns.boxplot(data=df, x='Activity Level', y='App Sessions')
plt.title('App Sessions by Activity Level')

plt.subplot(2, 3, 4)
sns.scatterplot(data=df, x='Distance Travelled (km)', y='Calories Burned')
plt.title('Calories Burned vs Distance Travelled')

plt.subplot(2, 3, 5)
sns.boxplot(data=df, x='Location', y='App Sessions')
plt.title('App Sessions by Location')

plt.tight_layout()
plt.show()
```

```python
# Step 5: Correlation Analysis
# Encoding categorical variables
df_encoded = pd.get_dummies(df, columns=['Gender', 'Activity Level', 'Location'], drop_first=True)

correlation_matrix = df_encoded.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
